```python
import mysql.connector
import pandas as pd

conn = mysql.connector.connect(
host='localhost',
user='root',
password='herobrinetxt',
database='lab4_db'
)
```

```python
#Q1. Fetch the customer details based on their language and age group.
lang = "en"
age = 18
query = f"SELECT * FROM customer_detail WHERE customer_lang = \"{lang}\" AND TIMESTAMPDIFF(YEAR, customer_dob, CURDATE()) >= {age};"
df = pd.read_sql(query,conn)

print(df)
```

```python
#Q2. Calculate the ratio of male to female customers and store the data in a CSV file.
query = "SELECT SUM(customer_gender='M') AS male_count,SUM(customer_gender='F') AS female_count,SUM(customer_gender='M')/SUM(customer_gender = 'F') AS male_female_ratio FROM customer_detail;"
df = pd.read_sql(query,conn)
df.to_csv('gender_ratio_IF_L10.csv', index=False)
```

```python
#Q3. Analyse the distribution of languages spoken by customers in different countries and store the data in a CSV file.
query = "SELECT customer_country,customer_lang,COUNT(*) AS language_count FROM customer_detail GROUP BY customer_country,customer_lang ORDER BY customer_country,language_count DESC;"
df = pd.read_sql(query,conn)
df.to_csv('lang_country_stats_IF_L10.csv', index=False)
```
